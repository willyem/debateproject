```python
txt_guns = open('constitutional_series/guns.txt')
all_text_guns = txt_guns.read()
newlist_guns = all_text_guns.split('\n')
newlist_guns = [x.strip() for x in newlist_guns if (x != " ") and (x != "")]
for_motion = []
against_motion = []
moderator = []

for line in newlist_guns: 
    if line.lower().startswith("for the motion:"):
        line = line.replace(" and ", ":")
        line = line.replace(",",":")
        for item in line.split(":")[1:]:
            for_motion.append(item.strip()+":")
        
for line in newlist_guns: 
    if line.lower().startswith("against the motion:"): 
        line = line.replace(" and ", ":")
        line = line.replace(",",":")
        for item in line.split(":")[1:]:
            against_motion.append(item.strip()+":")
            
for line in newlist_guns: 
    if line.lower().startswith("moderator"): 
        moderator.append(line.split(":")[1].strip()+":")

print for_motion
```

```python
    
#pseudocode: 
#go through each line of code and see if it has speaker X in it 
#if it does, remember the index 
#if the line has ANOTHER speaker in it, pass all lines between
#previous saved index and current index to a new list 

robert_baer_text = []
started = False
        
```

```python
allfolks = for_motion+against_motion+moderator

all_text_in_dict = {}
against_and_moderator = against_motion+moderator

for speaker in for_motion:
    
    all_text_in_dict[speaker] = []
    

    for index, line in enumerate(newlist_guns):
    
        if speaker in line and (started==False): 
            start_index = index
            started = True
                
        if started == True: 
            for anti in against_and_moderator: 
                if anti in line: 
                    all_text_in_dict[speaker].append(newlist_guns[start_index:index])
                    started = False
```
